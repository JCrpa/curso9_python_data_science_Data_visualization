.set_index()

Vamos mudar o índice de coluna e vamos utilizar inplace = True para definir que a mudança ocorrerá no próprio dataframe e que não é para criar um novo dataframe

```python
df_compara.set_index('País', inplace=True)
```

Vamos criar uma variável anos para trabalhar com todo o intervalo (range) de anos do dataframe que vai de 1980 a 2014

```python
anos_compara = list(map(str, range(1980, 2014)))
```

```python
anos_compara
```

Agora vamos delimitar através de .loc que queremos apenas os dados do Brasil

```python
Brasil = df_compara.loc['Brasil', anos_compara].astype(int)
Argentina = df_compara.loc['Argentina', anos_compara].astype(int)
compara = pd.DataFrame({'Brasil': Brasil, 'Argentina': Argentina}, index=anos_compara)
```

## Módulo 1 - aula 6 - Criando o primeiro gráfico com Matplotlib

```python
import matplotlib.pyplot as plt
```

### plt.xlabel() / plt.ylabel()

Adiciona informação de rótulo dos eixos

```python
compara.plot(figsize=(8, 4))
plt.title('Comparação da imigração entre Brasil e Argentina para no Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(rotation=45)
plt.show()
```
