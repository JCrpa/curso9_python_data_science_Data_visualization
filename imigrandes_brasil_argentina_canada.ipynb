## Módulo 5 - Gráficos interativos com Plotly - aula 08 - Desafio: criando uma animação para comparar diferentes dados

Tarefas:

- Reestruturar os Dados: Vamos preparar os dados para a animação.
- Criar a Figura Inicial: Começaremos com um gráfico mostrando apenas o primeiro ano (1980) para todos os 4 países.
- Gerar os "Frames": Cada ano subsequente será um "frame" (quadro) da animação, adicionando um novo ponto de dado a cada uma das 4 linhas.
- Adicionar Controles: Incluiremos um botão de "Play/Pause" e uma barra de rolagem (slider) para navegar pela animação.

```python
import pandas as pd

# Carregar os dados
df_compara = pd.read_csv('C:\\Users\\joaoc\\OneDrive\\Documentos\\Aprendizado_ONE\\formacao-tech-foundation\\curso9_python_data_science_Data_visualization\\imigrantes_canada.csv')
df_compara.set_index('País', inplace=True)

# Selecionar colunas de anos
anos_compara = [col for col in df_compara.columns if col.isdigit() and 1980 <= int(col) <= 2013]

# Dados por país, convertendo para int
Brasil = pd.to_numeric(df_compara.loc['Brasil', anos_compara], errors='coerce').fillna(0).astype(int)
Argentina = pd.to_numeric(df_compara.loc['Argentina', anos_compara], errors='coerce').fillna(0).astype(int)
Peru = pd.to_numeric(df_compara.loc['Peru', anos_compara], errors='coerce').fillna(0).astype(int)
Colombia = pd.to_numeric(df_compara.loc['Colômbia', anos_compara], errors='coerce').fillna(0).astype(int)

compara = pd.DataFrame({
    'Brasil': Brasil,
    'Argentina': Argentina,
    'Peru': Peru,
    'Colômbia': Colombia
}, index=anos_compara)
```

```python
import plotly.graph_objs as go

# Converta os índices para int para evitar confusão na animação
anos_int = list(map(int, anos_compara))

# Atualize os dados para usar anos_int ao invés de strings
frames = [
    go.Frame(
        data=[
            go.Scatter(x=anos_int[:i+1], y=Brasil.iloc[:i+1], name='Brasil', mode='lines+markers', line=dict(width=4)),
            go.Scatter(x=anos_int[:i+1], y=Argentina.iloc[:i+1], name='Argentina', mode='lines+markers', line=dict(width=4)),
            go.Scatter(x=anos_int[:i+1], y=Peru.iloc[:i+1], name='Peru', mode='lines+markers', line=dict(width=4)),
            go.Scatter(x=anos_int[:i+1], y=Colombia.iloc[:i+1], name='Colômbia', mode='lines+markers', line=dict(width=4)),
        ],
        name=str(i)
    )
    for i in range(len(compara))
]

# Traços iniciais também com anos_int e modo 'lines+markers'
fig = go.Figure(
    data=[
        go.Scatter(x=[anos_int[0]], y=[Brasil.iloc[0]], mode='lines+markers', name='Brasil', line=dict(width=4)),
        go.Scatter(x=[anos_int[0]], y=[Argentina.iloc[0]], mode='lines+markers', name='Argentina', line=dict(width=4)),
        go.Scatter(x=[anos_int[0]], y=[Peru.iloc[0]], mode='lines+markers', name='Peru', line=dict(width=4)),
        go.Scatter(x=[anos_int[0]], y=[Colombia.iloc[0]], mode='lines+markers', name='Colômbia', line=dict(width=4)),
    ],
    frames=frames
)

# Atualize o eixo x para numérico (int) também
fig.update_layout(
    title=dict(
        text='Imigração de países sul americanos para o\nCanadá no período de 1980 a 2013',
        y=0.95,
        x=0.5,
        xanchor='center',
        yanchor='top',
        font=dict(size=20)
    ),
    xaxis=dict(
        range=[1979, 2014],
        autorange=False,
        title='<b>Ano</b>',
        showgrid=True,
        gridcolor='lightgrey'  # grade leve no eixo x
    ),
    yaxis=dict(
        range=[0, 7000],  # Ajustado para 7000
        autorange=False,
        title='<b>Número de imigrantes</b>',
        showgrid=True,
        gridcolor='lightgrey'  # grade leve no eixo y
    ),
    updatemenus=[
        dict(
            type='buttons',
            showactive=False,
            x=-0.1,
            y=1.18,
            xanchor='left',
            yanchor='top',
            bgcolor='lightgreen',
            bordercolor='black',
            borderwidth=1,
            font=dict(size=14, color='black'),
            pad={'l': 10, 'r': 10, 't': 5, 'b': 5},
            buttons=[
                dict(
                    label='Play',
                    method='animate',
                    args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True, 'transition': {'duration': 0}}]
                )
            ]
        )
    ],
    sliders=[{
        'steps': [
            {
                'method': 'animate',
                'args': [[str(i)], {'frame': {'duration': 100, 'redraw': True}, 'mode': 'immediate'}],
                'label': str(anos_compara[i])
            } for i in range(len(frames))
        ],
        'transition': {'duration': 0},
        'x': 0,
        'y': -0.05,
        'currentvalue': {'prefix': 'Ano: ', 'font': {'size': 14}},
        'len': 1.0,
        'bgcolor': 'white',              # Fundo branco do slider
        'bordercolor': 'black',          # Borda preta
        'borderwidth': 1,
        'tickcolor': 'black',            # Cor dos ticks (preto suave)
    }],
    width=1000,
    height=600,
    margin=dict(l=20, r=20, t=120, b=20),  # Aumentei o topo pra caber o título com quebra
    font=dict(size=12)
)

fig.write_html("imigracao_america_sul_grafico_animado.html")
fig.show()
```
